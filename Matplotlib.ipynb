Crea un gráfico de líneas a partir de un DataFrame

```python
import pandas as pd
import matplotlib.pyplot as plt

df = {
    'Año': [2015, 2016, 2017, 2018, 2019],
    'Ventas': [200, 240, 300, 330, 400]
}

df = pd.DataFrame(df)

# Crear el gráfico de líneas
plt.figure(figsize=(8, 6))
plt.plot(df['Año'], df['Ventas'], marker='o')

# Añadir etiquetas y título
plt.title('Ventas Anuales')
plt.xlabel('Año')
plt.ylabel('Ventas')

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

crear un grafico de barras

```python
plt.figure(figsize=(8, 6))
plt.bar(df['Año'], df['Ventas'], color='skyblue')

# Añadir etiquetas y título
plt.title('Ventas Anuales')
plt.xlabel('Año')
plt.ylabel('Ventas')

plt.grid()
plt.show()
```

añadir gastos al df

```python
df = {
    'Año': [2015, 2016, 2017, 2018, 2019],
    'Ventas': [200, 240, 300, 330, 400],
    'Gastos': [150, 180, 210, 220, 300]
}

df = pd.DataFrame(df)

plt.figure(figsize=(8, 6))
plt.scatter(df['Ventas'], df['Gastos'], color='green', marker='o')

plt.title('Comparación de Ventas vs. Gastos')
plt.xlabel('Ventas')
plt.ylabel('Gastos')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Subplot 1: Gráfico de líneas
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas
plt.plot(df['Año'], df['Ventas'], marker='o', label='Ventas', color='blue')
plt.plot(df['Año'], df['Gastos'], marker='o', label='Gastos', color='red')
plt.title('Ventas y Gastos Anuales - Línea')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.legend()
plt.grid(True)

# Subplot 2: Gráfico de barras
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas
bar_width = 0.35  # Anchura de las barras
index = df['Año']

plt.bar(index - bar_width / 2, df['Ventas'], bar_width, label='Ventas', color='blue')
plt.bar(index + bar_width / 2, df['Gastos'], bar_width, label='Gastos', color='red')
plt.title('Ventas y Gastos Anuales - Barras')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.xticks(df['Año'])
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

crear un histograma para una columna del df

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Ventas'], bins=3, color='skyblue', edgecolor='black')
plt.title('Distribución de Ventas')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')

plt.grid(True)
plt.show()
```

Crear un gráfico de líneas que muestre varias columnas del DataFrame al mismo tiempo.

```python
plt.figure(figsize=(10, 6))

plt.plot(df['Año'], df['Ventas'], marker='o', label='Ventas', color='blue')

plt.plot(df['Año'], df['Gastos'], marker='o', label='Gastos', color='red')

plt.title('Comparación de Ventas y Gastos Anuales')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.legend()  # Añadir leyenda para diferenciar las líneas
plt.grid(True)  # Añadir cuadrícula para mejor visualización

plt.show()
```

Crear un gráfico de torta (pie chart) para visualizar las ventas como porcentaje del total.

```python
# Calcular el total de ventas
total_ventas = df['Ventas'].sum()

# Crear el gráfico
plt.figure(figsize=(8, 8))
plt.pie(df['Ventas'], labels=df['Año'], autopct='%1.1f%%', startangle=140)

plt.title('Distribución de Ventas por Año como Porcentaje del Total')

plt.axis('equal')  # Asegura que el gráfico sea un círculo
plt.show()
```
